+ (2 pts). Prove that $\|\cdot\|^{'}$ is a norm.

$$
\|\alpha x\|' = \sup_{\|y\|=1}|y^*\alpha x| = \sup_{\|y\|=1}|\alpha||y^* x| = |\alpha|\|x\|'
$$

$$
\|x + y\|' = \sup_{\|z\|=1}|z^*(x + y)| \leq \sup_{\|z\|=1}|z^* x||z^* y| = \|x\|'+\|y\|'
$$

Here we used the triangle inequality for absolute: $|x+y|\leq |x|+|y|$

$$
\|x\|' = 0 = \sup_{\|y\| = 1}|y^* x|
$$

As the supremum of absolute is zero, this means that every value is zero.

+ (2 pts). Find dual norms for Chebyshev norm $\left\|x\right\|_{\infty} = \max\limits_{i=1,\dots, N}|x_{i}|$ and $l_1$ norm $\left\|x\right\|_1 = \sum\limits_{i=1}^{N}|x_{i}|$.

First part
$$
\|x\|_{\infty} = \max_{i} |x_i|
$$

$$
\|x\|'_{\infty} = \sup_{\|y\|_{\infty}=1} |y^* x|
$$

As $\|y\|_{\infty} = 1$, this means that $|y_i| \leq 1$. Using the triangle inequality for absolute:

$$
|y^* x| = |\sum_{i = 1}^N y_i^*x_i| \leq \sum_{i = 1}^N |y_i^*x_i| = \sum_{i = 1}^N |y_i^*||x_i| \leq \sum_{i = 1}^N |x_i|
$$

Let's represent each $x_j$ as $|x_j|e^{i\varphi_j}$. Let's take $y_j = e^{i\varphi_j}$. $|y_j| = 1$,  so $\|y\|_{\infty} = 1$. Let's compute $|y^*x| = |\sum_{j}e^{-i\varphi_j}\cdot |x_j|e^{i\varphi_j}| = |\sum_{j}|x_j|| = \sum_{j}|x_j| = \|x\|_1$. The dual norm for $\|\cdot\|_\infty$ is $\|\cdot\|_1$.


Now second part
$$
\|x\|_{1} = \sum_{i} |x_i|
$$

$$
\|x\|'_{1} = \sup_{\|y\|_{1}=1} |y^* x|
$$

$$
|y^* x| \leq \sum_j|y_j^*||x_j|
$$

Let's prove $\sup_{\|y\|_1=1}|y^*x| \leq \max_i|x_i|$

There are two possibilities: $\max_j|x_j| = 0$ and $\max_j|x_j| \neq 0$. If $\max_j|x_j| = 0$, then $x = 0$, this is trivial. Otherwise, when $\max_j|x_j| \neq 0$, let's label $M = \max_j|x_j|$. $\frac{|x_j|}{M} \leq 1$.

$$
\sum_j|y_j^*||x_j| = M \sum_j |y_j|\frac{|x_j|}{M} \leq M \sum_j|y_j^*| = M = \max_j |x_j|
$$

To reach this bound, let's take $y = (0, ..., 0, 1, 0, ..., 0)^T$, the vector with only non-zero element with the same element as the maximum absolute in $x$, i.e. $y_i = 1 \Leftrightarrow |x_i| = max_{j}|x_j|$. $\|y\|_1 = 1$, $|y^* x| = |x_i| = max_{j}|x_j| = \|x\|_{\infty}$

So, the dual for $\|\cdot\|_1$ is $\|\cdot\|_{\infty}$

+ (2 pts). Find dual norm for $l_p$ norm $\left\|x\right\|_{p} = \left(\sum\limits_{i=1}^{N}|x_{i}|^{p}\right)^{1/p}$.

Let's use Young's inequality, that is $|ab|\leq \frac{|a|^p}{p}+\frac{|b|^q}{q}$, with $\frac{1}{p}+\frac{1}{q}=1$ and $p > 1$, $q > 1$. 

We want to prove, that $|y^*x| \leq \|y\|_p\|x\|_q$ when $\frac{1}{p}+\frac{1}{q}=1$ and $p > 1$, $q > 1$. If $x=0$ or $y=0$, this is trivial. Let's assume both are non-zero. Define $a = \frac{y}{\|y\|_p}$, $b = \frac{x}{\|x\|_q}$. $\|a\|_p = \frac{\|y\|_p}{\|y\|_p} = 1$, $\|b\|_q = 1$.

$$
|a^*b| = |\sum_i a_i^*b_i| \leq \sum_i |a_i||b_i| \leq \sum_i \frac{|a_i|^p}{p} + \frac{|b_i|^q}{q} = \frac{1}{p}\|a\|_p^p+\frac{1}{q}\|b\|_q^q = \frac{1}{p}+\frac{1}{q}=1
$$

Substituting $x$ and $y$ back, we get 

$$
\frac{|y^*x|}{\|y\|_p\|x\|_q} \leq 1
$$

$$
|y^*x|\leq \|y\|_p\|x\|_q
$$

So, back to dual norms (using the inequality above)

$$
\|x\|_p' = \sup_{\|y\|_p = 1}|y^*x| \leq \|x\|_q
$$

To reach this bound, we should take $z_j = e^{i\varphi_j}|x_j|^{q-1}$ and $y = \frac{z}{\|z\|_p}$ (we represent $x_j = |x_j|e^{i\varphi_j}$). Indeed, $\|y\|_p = 1$, and $\|z\|_p=\left(\sum_j|x_j|^{p(q-1)}\right)^{1/p}=\left(\sum_j|x_j|^{q}\right)^{(q-1)/q}=\|x\|_q^{q-1}$ (here we used condition for $p$ and $q$). 

$$
|y^*x| = \frac{\sum_j |x_j|^{q-1}e^{-i\varphi_j} |x_j|e^{i\varphi_j}}{\|z\|_p} = \frac{\sum_j |x_j|^{q}}{\|z\|_p} =\frac{\|x\|^{q}_q}{\|x\|_q^{q-1}}=\|x\|_q
$$


+ (2 pts). Find dual norm for weighted norm $\left\|x\right\|_{A} = \left(x^T A x\right)^{1/2}$ with positive-definite matrix $A$ and $x\in\mathbb{R}^{N}$.

$$
\|x\|_A = \sqrt{x^TAx}
$$

We can represent $A$ as sum of symmetric and anti-symmetric parts: $A=\frac{A+A^T}{2}+\frac{A-A^T}{2}$. So, $x^TAx = x^T(\frac{A+A^T}{2}+\frac{A-A^T}{2})x=x^T(\frac{A+A^T}{2})x$, since it is a number, so its transpose is the same. Denote $\frac{A+A^T}{2}=B$. We found that

$$
\|x\|_A = \sqrt{\frac{x^T(A+A^T)x}{2}} = \sqrt{x^TBx} = \|x\|_B
$$

$$
\|x\|_A' = \sup_{\|y\|_B=1}|y^Tx|
$$

$$
|y^Tx|=|y^TBB^{-1}x| = |(B^Ty)B^{-1}(x)| = |\langle B^Ty,x \rangle|
$$

Where $\langle u, v \rangle = u^T B^{-1} v$ - inner product. It satisfies all inner product properties, positiveness and symmetry in particular, because $B^{-1}$ is positive definite too ($A+A^{-1}$ is positive definite, so is inverse). Thus, we can use Cauchy-Schwarz inequality

$$
|\langle B^Ty, x \rangle|^2 \leq \langle B^Ty, B^Ty \rangle \langle x, x \rangle
$$

$$
\langle B^Ty, B^Ty \rangle = y^TBB^{-1}B^Ty=y^TB^Ty=y^TBy
$$

The last equality is true, as $B=B^t$ by construction.

$$
\langle x, x \rangle = x^TB^{-1}x
$$

Collecting everything together, we get

$$
|y^Tx| \leq \|y\|_B\|x\|_{B^{-1}}
$$

Consequently,

$$
\sup_{\|y\|_B=1}|y^Tx| \leq \|x\|_{B^{-1}}
$$

To reach the bound, we can take $y^T=\frac{x^TB^{-1}}{\|x\|_{B^{-1}}}$. 

$$
\|y\|_B = \sqrt{y^TBy} = \sqrt{\frac{x^TB^{-1}BB^{-T}x}{\|x\|_{B^{-1}}^2}}=\frac{\sqrt{x^TB^{-T}x}}{\|x\|_{B^{-1}}}=\frac{\sqrt{x^TB^{-1}x}}{\|x\|_{B^{-1}}}=1
$$

Where $B^{-T}=(B^{-1})^T=(B^{T})^{-1}$

So, 

$$
|y^Tx|=\left|\frac{x^TB^{-1}x}{\|x\|_{B^{-1}}}\right|=\frac{\|x\|_{B^{-1}}^2}{\|x\|_{B^{-1}}}=\|x\|_{B^{-1}}=\|x\|_{(\frac{A+A^T}{2})^{-1}}
$$

Denoting $u_i$ as i-th column of matrix $U$ and $v_i$ as i-th column of matrix $V$, we can rewrite matrix $A$:

$$
A = \sum_{i=1}^r u_iv_i^T

$$

Last identity follows from

+ (5 pts). Propose an algorithm for the computation of $\Vert B \Vert^2_F$ and estimate its complexity with respect to $n, m, r$.

$$

$$
\|B\|_F^2 = \Tr(B^TB) = \Tr\left( \sum_{i=1}^r\sum_{j=1}^r\sum_{k=1}^r\sum_{l=1}^r (v_k\circ v_l)(u_k\circ u_l)^T(u_i\circ u_j)(v_i\circ v_j)^T \right)
$$

Since the trace of sum is sum of traces, we can rewrite it as:

$$
\|B\|_F^2 = \sum_{i,j,k,l=1}^r \Tr\left((v_k\circ v_l)(u_k\circ u_l)^T(u_i\circ u_j)(v_i\circ v_j)^T \right)

Under the trace we can exchange matrices following this rule $\Tr(AB)=\Tr(BA)$.

\|B\|_F^2 = \sum_{i,j,k,l=1}^r \Tr\left((u_k\circ u_l)^T(u_i\circ u_j)(v_i\circ v_j)^T(v_k\circ v_l) \right)

$u$-factors result into the number, so do the $v$-factors. So, we can get rid of $\Tr$.

$$
\|B\|_F^2 = \sum_{i,j,k,l=1}^r \left((u_k\circ u_l)^T(u_i\circ u_j)\right) \left((v_i\circ v_j)^T(v_k\circ v_l)\right)

To compute $u_i\circ u_j$, we need $O(n)$ operations (n multiplications). To compute $(u_k\circ u_l)^T(u_i\circ u_j)$ we need $O(n)$ operations (n multiplications and n-1 additions). $$

Again, assume $K > 2NLe$ (again, need exact - compute numerically, here is just an approximation). Then 

$$
\frac{M}{1-\frac{NLe}{K}}\left(\frac{NLe}{K}\right)^K < 2M\left(\frac{1}{2}\right)^K
$$

So, to construct a matrix X, we should do Taylor series up to order K, which can be found from this inequality

$$
2M\left(\frac{1}{2}\right)^K < \varepsilon
$$

If $K > \max\left\{ 2NLe, \frac{\log\frac{2M}{\varepsilon}}{\log2} \right\}$

From this inequality we can find the needed order of approximation. To build the matrix $X$ we approximate every element of $A$ with first $K$ terms of Taylor series. To bound the rank we can represent $X$ as sum of rank-1 matrices, as in the previous task. But now we will have $N^K$ terms, it can be seen from the first equation in this block. So, the rank doesn't exceed $N^K\approx N^{NL}$.

```python
import numpy as np
from numpy.linalg import svd
from numpy.random import rand
from numpy.linalg import matrix_rank as rank
import jax.numpy as jnp
from math import factorial
import matplotlib.pyplot as plt
```

$\cos(xy)$ - all derivatives are less than 1 (because $x$, $y$ are in $[0,1]$), $M = 1$

```python
def calculate_matrix(order, x, y):
  answer = np.zeros((x.shape[0], y.shape[1]))
  for i in range(0, order + 1, 2):
    if i % 2:
      answer -= (x ** i) @ (y ** i) / factorial(i)
    else:
      answer += (x ** i) @ (y ** i) / factorial(i)
  return rank(answer)

def find_order(eps):
  order = 3
  while (np.e / order) ** order > eps * (1 - np.e / order):
    order += 1
  return order
```

Since the $\|\cdot\|_2$ norm preserves distances, let's compute the distances between atoms and store them in some order, this will identify the unique (almost) structure of the molecule. We can turn this array into a tuple and make this vector of lengths a 'key'. Due to machine precision, we will call the molecules the same, if the elements of such vectors rounded to $10^{-10}$ are the same.

```python
from numpy.linalg import norm
import numpy as np
```

```python
def distinguish_molecule(data):
    dictionary = {}
    for index in np.arange(data.shape[0]):
        coordinates = []
        for i in range(len(data[index]) - 1):
            for j in range(i + 1, len(data[index])):
                coordinates.append([data[index][i, 0] - data[index][j, 0], data[index][i, 1] - data[index][j, 1]])
        norms = []
        for i in range(len(coordinates)):
            norms.append(round(norm(coordinates[i], 2), 10))
        for i in range(len(coordinates) - 1):
            norms.append(round(coordinates[i][0] * coordinates[i + 1][0] + coordinates[i][1] * coordinates[i + 1][1], 5))
        norms = tuple(norms)
        if norms not in dictionary:
            dictionary[norms] = []
        dictionary[norms].append(index)
    return dictionary
```

+ (7 pts). Learn about [Kabsch algorithm](https://en.wikipedia.org/wiki/Kabsch_algorithm). Implement it and run on the database. Produce a list of distinct molecules and visualise them.

I implement 'improved' Kabsch algorithm, which rotates the molecules and tells, whether they are the same or not. It does rotation, and then checks the frobenius norm of difference of matrix of molecule coordinates. If the difference is less than $10^{-10}$, the molecules are considered same. 

```python
from numpy.linalg import svd
```

```python
def kabsch(coord1, coord2):
    H = coord1.T @ coord2
    U, S, VT = svd(H)
    orientation = np.sign(np.linalg.det(U @ VT))
    R = U @ np.diag(np.array([1, orientation])) @ VT
    new_pos = np.array([R @ pos for pos in coord2])
    norm_2 = norm(coord1 - new_pos, 'fro')
    return norm_2
    
```

```python
used = [0]
types = {0 : [0]}

while len(used) < len(molecules):
    for i in range(len(molecules)):
        if i not in used:
            used.append(i)
            flag = True
            for key in types:
                if kabsch(molecules[i], molecules[key]) < 1e-5:
                    types[key].append(i)
                    flag = False
            if flag:
                types[i] = [i]
    
count = 0
for key in types:
    count += len(types[key])
    print(len(types[key]))
    print(types[key])
print(count)
                    
```

## Brief intro

The main component of [GAN](https://pytorch.org/tutorials/beginner/dcgan_faces_tutorial.html) is Generator network. Generator is such model that takes as input random vector and produces an image. In this task, we will work with the already pre-trained generator $G$. Here are a few things you need to know about this model:



1.   **Model input:** batch of random vectors $z\sim N(0,1)$ with normally distributed values of size **(Batch_Size,100)**. Here, Batch_Size is a number of images we want to generate
2.   **Model output:** $\mathrm{images} = G(z)$ - batch of images with size **(Batch_Size,3,64,64)**. Here **3** - defines number of image channels (3 for RGB images), **64, 64** - width and height.
3.   **Model structure:** generator is a set of sequentially appplied functions. Each function input is the output of the previous one. In other words, $G(z) = f_N(f_{N-1}...f_1(f_0(z)))$. $N$ - defines number of functions in generator.



### Disentanglement

The good property of GAN is that generator input space has a set of meaningful directions - moving input vector along this directions produce interpretable changes in image. For example, we can change a Pose, Age, Expression or even add eyeglasses. You can read more about it on [resource](https://genforce.github.io/interfacegan/). The main question is how to find such good directions. This problem is called disentanglement.

![](https://github.com/genforce/interfacegan/blob/master/docs/assets/teaser.jpg?raw=true)

## Problem discription 

We are going to find meaningful directions using SVD applied over the Jacobian matrix of some hidden output with respect to model input. **i-th** hidden output of generator model is defined by formula $G_{i}(z) = f_i(f_{i-1}...f_1(f_0(z))))$, where $i < N$. So, the task could be splitted on four main parts:

1. **(4 pts)**   Write the function **Get_Hidden_Output**, which reproduces formula $G_{i}(z) = f_i(f_{i-1}...f_1(f_0(z))))$
2. **(3 pts)**   Calculate jacobian matrix $J$ of generator $G$ for several **i** values. For each of them:

*   Calculate the jacobian matrix with respect to $z$: $\frac{\partial G(z)}{\partial z}$
*   Reshape jacobian matrix to have the size **(Batch_Size * Hidden_Output_Size,100)**

3. **(3 pts)**  Apply SVD to jacobian matrix ($J = U @ S @ V.T$) from previous step

```python
class Generator(nn.Module):
    def __init__(self, ngpu):
        super(Generator, self).__init__()
        self.ngpu = ngpu
        self.main = nn.Sequential(
            # input is Z, going into a convolution
            nn.ConvTranspose2d( nz, ngf * 8, 4, 1, 0, bias=False),
            nn.BatchNorm2d(ngf * 8),
            nn.ReLU(True),
            # state size. (ngf*8) x 4 x 4
            nn.ConvTranspose2d(ngf * 8, ngf * 4, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ngf * 4),
            nn.ReLU(True),
            # state size. (ngf*4) x 8 x 8
            nn.ConvTranspose2d( ngf * 4, ngf * 2, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ngf * 2),
            nn.ReLU(True),
            # state size. (ngf*2) x 16 x 16
            nn.ConvTranspose2d( ngf * 2, ngf, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ngf),
            nn.ReLU(True),
            # state size. (ngf) x 32 x 32
            nn.ConvTranspose2d( ngf, nc, 4, 2, 1, bias=False),
            nn.Tanh()
            # state size. (nc) x 64 x 64
        )
    def get_function_output(self, inp, index=0):
      if index == 0:
        inp = inp[:,:,None,None]
      return self.main[index](inp)
    def get_number_of_layers(self):
      return len(self.main)

    def forward(self, input):
        return self.main(input[:,:,None,None])
```
